```python
# Basic libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("application_test.csv")
```

```python
# data check
df.head()
df.shape
df.info()
df.describe()
```

```python
# data cleaning

#remove unnecessary column
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
print(df)
# handle missing values
df.isnull().sum().sort_values(ascending=False).head(10)
# keep columns with 50% data
threshold = len(df) * 0.5
df = df.dropna(thresh=threshold, axis=1)
# fill numerical values
df.fillna(df.median(numeric_only=True), inplace=True)
# fill categorical values
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
# FEATURE ENGG

#loan to income ratio (high ratio, high risk)
df['loan_to_income_ratio'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']

#EMI Burden
df['EMI_ratio'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']



```

```python
# EDA

import seaborn as sns
sns.histplot(df['loan_to_income_ratio'], kde=True)
#Correlation Check
sns.heatmap(df.corr(numeric_only=True), annot=False)
```

Business Logic (JPMC style)

High-risk if:

Loan-to-income ratio > 0.5
EMI ratio > 0.3
Low income + high credit

```python
# DEFINE HIGH-RISK LOANS

df['High_Risk'] = (
    (df['loan_to_income_ratio'] > 0.5) |
    (df['EMI_ratio'] > 0.3)
)
print(df)
```

```python
# ANALYSIS OF HIGH-RISK CUSTOMERS

df['High_Risk'].value_counts()

df.groupby('High_Risk')['AMT_INCOME_TOTAL'].mean()
# insight : low income customers are of high risk

df.groupby('High_Risk')['AMT_CREDIT'].mean()
# insight : high risk customers take large loans
```

```python
# VISUAL ANALYSIS

plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='AMT_INCOME_TOTAL', y='AMT_CREDIT', hue='High_Risk', alpha=0.6)
plt.title('Credit Amount vs. Total Income, Colored by High Risk Status')
plt.xlabel('Total Income (AMT_INCOME_TOTAL)')
plt.ylabel('Credit Amount (AMT_CREDIT)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()



sns.boxplot(data=df, x='High_Risk', y='loan_to_income_ratio')
```

**INSIGHTS**
- High loan-to-income ratio strongly indicates risk
- High EMI ratio correlates with financial stress
- Low-income customers with high credit are high-risk

***Business Recommendations***

- Restrict loans above 50% income ratio
- Monitor high EMI customers
- Improve credit scoring for better risk control
